```python
#problem1,part1
def gcd(a,b):
    a=abs(a)
    b=abs(b)
    while b!=0:
        a,b=b,a%b
    return a
```

```python
#problem1,part2
gcd(-48,18)
```

```python
#problem2,part1
def euler_limit(n):
    return (1+1/n)**n
```

```python
#problem2,part3
import math
def euler_sum(k):
    t=0
    for i in range(k+1):
        t=t+1/math.factorial(i)
    return t
```

```python
#problem2,part4
def euler_approx(epsilon):
    t=0
    k=0
    while True:
        a=1/math.factorial(k)
        t=t+a
        if abs(t-math.e)<epsilon:
            return t
        k=k+1
```

```python
#problem2,part5
def print_euler_sum_error(n):
    approx=euler_sum(n)
    error=approx-math.e
    print("Euler Sum Approximation:",approx)
    print("Error:",error)
def print_euler_limit_error(n):
    approx=euler_limit(n)
    error=approx-math.e
    print("Euler Limit Approximation:",approx)
    print("Error:",error)
```

```python
print_euler_sum_error(1000)
print_euler_limit_error(1000)
```

```python
#problem3,part1
def is_power_of_2(n):
    while n>1:
        if n%2!=0:
            return False
        n=n//2
    return True
```

```python
#problem3,part2
def is_power(b,n):
    while n>1:
        if n%b!=0:
            return False
        n=n//b
    return True
```
